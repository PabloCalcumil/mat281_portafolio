El **cáncer de mama**  es una proliferación maligna de las células epiteliales que revisten los conductos o lobulillos mamarios. Es una enfermedad clonal; donde una célula individual producto de una serie de mutaciones somáticas o de línea germinal adquiere la capacidad de dividirse sin control ni orden, haciendo que se reproduzca hasta formar un tumor. El tumor resultante, que comienza como anomalía leve, pasa a ser grave, invade tejidos vecinos y, finalmente, se propaga a otras partes del cuerpo.

El conjunto de datos se denomina `BC.csv`, el cual contine la información de distintos pacientes con tumosres (benignos o malignos) y algunas características del mismo.


Las características se calculan a partir de una imagen digitalizada de un aspirado con aguja fina (FNA) de una masa mamaria. Describen las características de los núcleos celulares presentes en la imagen.
Los detalles se puede encontrar en [K. P. Bennett and O. L. Mangasarian: "Robust Linear Programming Discrimination of Two Linearly Inseparable Sets", Optimization Methods and Software 1, 1992, 23-34].


Lo primero será cargar el conjunto de datos:

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA


%matplotlib inline
sns.set_palette("deep", desat=.6)
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
# cargar datos
df = pd.read_csv(os.path.join("data","BC.csv"), sep=",")
df['diagnosis'] = df['diagnosis'] .replace({'M':1,'B':0})
df.head()
```

Basado en la información presentada responda las siguientes preguntas:

#### 1 )  Normalizar para las columnas numéricas con procesamiento **StandardScaler**.

```python
#Notar que el id no es necesario normalizarlo, dado que no aporta informacion de los tumores
# solo sirve para diferenciar a los pacientes
target = df['diagnosis']
df = df.drop(['id', 'diagnosis'], axis = 1) 
transform = StandardScaler().fit_transform(df)

#Dataframe normalizado
df_normalizado = pd.DataFrame(transform, columns = df.columns)
df_normalizado
```

#### 2) Realice un gráfico de correlación. Identifique la existencia de colinealidad.

```python
fig = plt.figure(figsize = (25,25))
sns.heatmap(df_normalizado.corr().round(2), #Redondeo
            annot = True, fmt = 'g',        #Valores
            cmap = 'Blues')                 #Color
plt.show()
```

__Respuesta:__ Es facil notar que existe cierta colinealidad, esto dado por el tono de los azules, en donde los mas oscuros son valores que están cercanos a 1, sin contar obviamente los de la diagonal.

#### 3) Realizar un ajuste PCA con **n_components = 10**. Realice un gráfico de la varianza y varianza acumulada. Interprete.

```python
#PCA con 10 componentes
pca = PCA(n_components = 10)
princ_comp = pca.fit_transform(df_normalizado.values)

#Varianzas
varianzas = np.round(pca.explained_variance_ratio_* 100, decimals = 2)
columns = ['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7', 'PC8', 'PC9', 'PC10']

#Grafico de barras
plt.figure(figsize = (16,6))
plt.bar(x = range(1,11),      #Eje x
        height = varianzas,   #Varianzas
        tick_label = columns) #Etiqueta por barra (PC numero)

#Etiquetas
plt.ylabel('Percentate of Variance Explained')
plt.xlabel('Principal Component')
plt.title('PCA Scree Plot')
plt.show()
```

```python
#Varianzas acumuladas
varianzas_acum = np.cumsum(varianzas)

#Etiquetas PCA acumulados
for i in range(1, 10):
    columns[i] = columns[0] + '+...+' + columns[i]

#Grafico
plt.figure(figsize=(25,25))
plt.bar(x = range(1,11),          #Eje x
        height = varianzas_acum,  #Valores
        tick_label = columns)     #Etiquetas eje x

#Etiquetas
plt.ylabel('Percentate of Variance Explained')
plt.xlabel('Principal Component Cumsum')
plt.title('PCA Scree Plot')
plt.show()
```

__Respuesta:__ Se puede notar que en el primer gráfico, el `PC1` es el que mayor varianza tiene, esto obtenido por el _PCA_, mientrás que los demás, en especial los $7, 8, 9$ y $10$, tienen valores muy bajos.

Por otro lado, en el segundo gráfico, se representa con la acumulada como las últimas barras son similares dado por lo expuesto anteriormente.

#### 4) Devuelva un dataframe con las componentes principales.

```python
columns = [col[-3] + col[-2] + col[-1] for col in columns]
df_princ = pd.DataFrame(data = princ_comp, columns = columns)
df_princ
```

#### 5) Aplique al menos tres modelos de clasificación. Para cada modelo, calcule el valor de sus métricas.

```python
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score, f1_score

#Función summary laboratorios anteriores
def summary_metrics(y_test,y_pred):
   
    # metrics

    print('\nMatriz de confusion:\n ')
    print(confusion_matrix(y_test,y_pred))

    print('\nMetricas:\n ')
    print('accuracy:   ',accuracy_score(y_test, y_pred))
    print('recall:     ',recall_score(y_test, y_pred))
    print('precision:  ',precision_score(y_test, y_pred))
    print('f-score:    ',f1_score(y_test, y_pred))
    print("")

    return
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression #Logistica
from sklearn.tree import DecisionTreeClassifier     #Arbol clasificacion
from sklearn.svm import SVC                         #SVD

#Entrenar modelo
x_train, x_test, y_train, y_test = train_test_split(df_princ, target, test_size=0.35, 
                                                    train_size=0.65,
                                                    random_state=1982)

#Modelo regresion log y prediccion
log_model = LogisticRegression(solver = 'liblinear').fit(x_train, y_train)
log_pred = log_model.predict(x_test)

#Metricas
summary_metrics(y_test, log_pred)
```

```python
#Modelo SVC y prediccion
svc_model = SVC()
svc_model.fit(x_train, y_train)
svc_pred = svc_model.predict(x_test)

#Metricas
summary_metrics(y_test, svc_pred)
```

```python
#Modelo de arbol de clasificación y prediccion
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
ac_pred = model.predict(x_test)

#Metricas
summary_metrics(y_test, ac_pred)
```
